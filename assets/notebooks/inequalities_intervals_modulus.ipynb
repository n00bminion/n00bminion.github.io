### <a id='toc1_1_1_'></a>[Inequalities, Intervals and Modulus](#toc0_)

Inequalities are relationships concerning the ordering of numbers. The available inequalities are: $$ <, \space >, \space \ge, \space \le $$

### <a id='toc1_1_2_'></a>[Manipulating inequalities](#toc0_)

These are quite simple rules and should be expected. Number 5 might is the only one that's worth noting

- If $x>y$ and $c$ is any real number, then $x+c>y+c$ also holds. We can add $c$ ($c$ can be both negative or positive) to both side of the inequality without the direction of the inequality changing. 
- If $x>y$ and $y>z$ then $x>z$
- If $x>y$ and $c>0$ then $xc>yc$. This just scales $x$ and $y$ by $c$ so inequality does not need to flip
- For any real number $x$, $x^2 \ge 0$
- If $x>y$ and $c<0$ then $xc<yc$. We must flip the inequality because we are scaling $x$ and $y$ by $c$ but in opposite directions now. Remember that $-x<-y$

### <a id='toc1_1_3_'></a>[Intervals](#toc0_)

A close interval of $\mathbb{R}$ is a set of the form $\{x \in \mathbb{R}: a \le x \le b\}$ where $a$ and $b$ are two reals. This can also written as $[a,b]$. 

A open interval of $\mathbb{R}$ is a set of the form $\{x \in \mathbb{R}: a < x < b\}$ where $a$ and $b$ are two reals. This can also written as $(a,b)$. 

You could also have am mix of close and open $[1, 4) = \{x \in \mathbb{R}: 1 \le x < 4 \}$

You could also have union ($\cup$) of intervals like this example $$\{x \in \mathbb{R}: x^2-7x+12 > 0 \}$$ $$ \Rightarrow (x-4)(x-3) > 0 $$

For this to be true, both $x-4$ and $x-3$ must positive or negative for it be larger than 0. 
- For both to be negative, $x$ must be smaller than 3
- For both to be positive, $x$ must be larger than 4

Therefore, we have $(-\infty,3) \cup (4, \infty)$  


### <a id='toc1_1_4_'></a>[Modulus](#toc0_)

For a given real number $x$, its modulus or absolute value, $|x|$ is the absolute value of $x$ from 0. $$ |-3| = |3| = 3  $$

We can also combine modulus with intervals and inequality like this $$\{x \in \mathbb{R}: |x+2| > 8 \}$$ 

With some simple mental maths, we see that:
- x must be larger than 6
- x must be smaller than -10 

Therefore the union intervals is:
$$ (-\infty,-10) \cup (6, \infty) $$
