### <a id='toc1_1_1_'></a>[Functions](#toc0_)

A function $ f : A \rightarrow B $ is a rule which associate each $x$ member of set A which we call **domain** of $f$ (we also say $f$ is defined on A) to a **unique** $y$ value in the set B which we call **codomain** of $f$.

We then write $$ y = f(x) $$

Where we call $f(x)$ the value of $f$ at $x$. The set of values of $f$ as $x$ takes all the values in A is called the range of $f$. The range of the $f$ is a subset of the codomain of $f$.

For some functions, the input might not make sense and/or is not defined. $f(x)= \tfrac{1}{x}$ is one where $x=0$ is impossible. We then say function is not defined at $x = 0$. Since the functions works for all real numbers, we can say that the "function $ f $ is defined by $ f(x) = \tfrac{1}{x}, x \ne 0$" or "function $$f : \mathbb{R}\backslash \{0\} \rightarrow \mathbb{R}$$ is defined by $f(x) = \tfrac{1}{x}$". 

Note that $$ \mathbb{R}\backslash \{0\} $$ denote real numbers excluding 0

<br>

### <a id='toc1_1_2_'></a>[Even & Odd Functions](#toc0_)

$f$ is an **even** function if $f(x) = f(-x)$

$f$ is an **odd** function if $f(x) = -f(-x)$

Note that all even functions have a symmetry about the $y$-axis (reflected by the $y$-axis) whereas the odd functions will have a rotational symmetry of order 2. This effectively mean you just have to rotate it by 90 degrees twice to get back to the same graph.

An example of even function is $ f(x) = x^2 $. Both positive and negative $x$ maps to the same positive $y$ value. For the odd function, we can use $ f(x) = x $, where positive $x$ maps to positive $y$ only and vice versa. This satisfies the condition above $f(x) = -f(-x)$.

<br>

### <a id='toc1_1_3_'></a>[Periodic Functions](#toc0_)

$f$ is a **periodic** function if $f(x) = f(x+a)$ where $a$ is the period of $f$.

Usually periodic functions are some sort of trignometric functions.

<br>

### <a id='toc1_1_4_'></a>[Polynomials Functions](#toc0_)

The general polynomial of degree $n$ in $x$ can be written as

$$ y = f(x) = a_nx^n + a_{n-1}x^{n-1} + a_{n-2}x^{n-2} + ... +  a_1x^1 + a_0 \nonumber $$

where $a_n, a_{n-2}, a_{n-3} ... a_1, a $ are numbers or coefficients and $a_n \ne 0$. The degree of the polynomial is the highest power of $x$.

A popular polynomial function is the **quadratic** function, where it is a polynomial of degree 2. Rearranging the polynomial equation can show the position and shape of the graph. $$ f(x) = ax^2 + bx + c \nonumber $$ 

Dividing all terms by $a$ and complete the square

$$ f(x) = a\left(x^2 + \frac{b}{a}x + \frac{c}{a}\right) \implies a\left(\left(x + \frac{b}{2a}\right)^2 - \frac{b^2}{4a^2} + \frac{c}{a}\right) \nonumber $$
 
Finally group the non $x$ terms in the bracket together

$$ a\left( \left(x + \frac{b}{2a}\right)^2 - \frac{b^2-4ac}{4a^2} \right) \nonumber $$


The graph of the $f(x)$ is thus obtained from $y=x^2$. This is shifted to the right by $-\frac{b}{2a}$, shifted downward by $\frac{b^2-4ac}{4a^2}$ and then scaled by $a$. Note that a negative value of $a$ will flip the graph upside down (reflection across the x axis) along with any scaling by $a$

### <a id='toc1_1_5_'></a>[Rational Functions](#toc0_)

Rational function $f$ is defined as

$$ f(x) = \frac{p(x)}{q(x)} \nonumber $$

where both $p$ and $q$ are polynomials. We need to be careful to exclude points where $q(x)$ is zero from the domain of both $p$ and $q$.

An example of this is we ignore 0 from all real values in the domain like below and the range of $f$ is R

$$ f: \mathbb{R} \backslash \{0\} \rightarrow \mathbb{R} \space defined \space by \space f(x) = \frac{1}{x} \nonumber$$

Or any value that will make the denominator equal to zero. In this case, we ignore 1 from the domain and the range of $f$ is R

$$ f: \mathbb{R} \backslash \{1\} \rightarrow \mathbb{R} \space defined \space by \space f(x) = \frac{x^2}{x^3-1} \nonumber$$

### <a id='toc1_1_6_'></a>[Trignometric Functions](#toc0_)

Before going into trignometric functions, we need to define radian and what it is. (This is a definition from the text book)

A radian is the angle subtended at the center of a circle by an arc which is the same length as the radius (subtended is a word I can never remember but it means opposite of something so in this context it is the angle at the center of of the circle which is opposite the arc, where the arc is a part of the circumference and this arc has the same length as the radius). If we have a unit circle where the radius is one, then the arc has a length of 1. 

We know formula of circumference of circle is $2\pi r$ which then equates to 360 degrees. Thus on a unit circle:

$$ 2\pi \space radian = 360^\circ \nonumber $$

$$ \pi \space radian = 180^\circ \nonumber $$

$$ \pi/2 \space radian = 90^\circ \nonumber $$

And in general, using the below, we can convert between radian and degree.

$$ 1 \space radian = \frac{180^\circ}{\pi}  \nonumber  $$

For example,  

$$ \frac{\pi}{3} \space radian \Rightarrow \frac{\pi}{3} . \frac{180^\circ}{\pi} = 60^\circ \nonumber $$

Trignometric functions are functions of angles which are measured in radians. If we see the below unit circle where angle $t$ and the point $(x,y)$ on the circumference of the circle. We can deduce that that $cos(t)= x$ and $sin(t) = y$. Using Pythagoras theorem, you can see that $$ cos^2(t) + sin^2(t) = 1 $$

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

fig, ax = plt.subplots()
# Move the left and bottom spines to x = 0 and y = 0, respectively.
ax.spines[["left", "bottom"]].set_position(("data", 0))
# Hide the top and right spines.
ax.spines[["top", "right"]].set_visible(False)

ax.add_patch(plt.Circle((0, 0), radius = 1, edgecolor='grey', facecolor='white'))
ax.add_patch(
    Arc(xy=(0,0), width=0.3, height=.3, theta1= 360, theta2= 50)
)

ax.tick_params(axis='both', which='major', labelsize=7.5)

plt.plot([0, 0.64], [0, 0.76], 'red',  lw=2,)
plt.plot([0, 0.64], [0, 0], 'black',  lw=2,)
plt.plot([0.64, 0.64], [0, 0.76], 'k-',  lw=2,)

plt.annotate('cos(t)', xy=(0.2,-0.1), fontsize=8.5, color='grey')
plt.annotate('sin(t)', xy=(0.7,0.35), fontsize=8.5, color='grey')
plt.annotate('1', xy=(0.27,0.45), fontsize=8.5, color='red')
plt.annotate('t', xy=(0.18,0.05), fontsize=8.5, color='black')
plt.annotate('(x,y)', xy=(0.64,0.78), fontsize=10, color='black')

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

ax.set_aspect('equal', adjustable='datalim')
ax.plot()   #Causes an autoscale update.

plt.show()
```

Along with this, we also have other trignometric identities:

$$ tan(t) = \frac{sin(t)}{cos(t)};\space cot(x) = \frac{cos(t)}{sin(t)};\space sec(x) =  \frac{1}{cos(t)};\space cosec(x) =  \frac{1}{sin(t)} \nonumber$$

Here's the visualisation of the different trignometric functions.

```python
import numpy as np
from matplotlib.ticker import FormatStrFormatter, MultipleLocator

x = np.linspace(-2 * np.pi, 2 * np.pi, 10000)
fig, ax = plt.subplots(2, 3, figsize=(17, 7))

def plot_trig(ax ,x, y, title):
    ax.set_title(title)
    ax.grid()
    ax.plot(
        x/np.pi, y, 'black'
    )
    ax.xaxis.set_major_formatter(FormatStrFormatter(r'%g $\pi$'))
    ax.xaxis.set_major_locator(MultipleLocator(base=0.5))


def bound_y(y):
    y[y>1000] = np.inf
    y[y<-1000] = np.inf
    return y

tan_y = bound_y(np.tan(x))
cosec_y = bound_y(1/np.sin(x))
sec_y = bound_y(1/np.cos(x))
cot_y = bound_y(1/tan_y)

charts = {
    'sin(t)':dict(ax=ax[0,0], x=x, y=np.sin(x)),
    'cos(t)':dict(ax=ax[0,1], x=x, y=np.cos(x)),
    'tan(t)':dict(ax=ax[0,2], x=x, y=tan_y),
    'cosec(t)':dict(ax=ax[1,0], x=x, y=cosec_y),
    'sec(t)':dict(ax=ax[1,1], x=x, y=sec_y),
    'cot(t)':dict(ax=ax[1,2], x=x, y=cot_y),
}

for k,v  in charts.items():
    plot_trig(v['ax'], v['x'], v['y'], k)

ax[0,2].set_ylim(-5,5)
ax[1,0].set_ylim(-5,5)
ax[1,1].set_ylim(-5,5)
ax[1,2].set_ylim(-5,5)

ax[1,0].yaxis.set_major_locator(MultipleLocator(base=1))
ax[1,1].yaxis.set_major_locator(MultipleLocator(base=1))


for i in range(0,2):
    for j in range(0,3):
        ax[i,j].spines[["top", "right"]].set_visible(False)
        ax[i,j].spines[["bottom", "left"]].set_position(("data", 0))

```

 A few observations here:

- Because $cos(t)$ has a symmetic about the y-axis, it is an even function whilst $sin(t)$ is odd as you rotate the graph twice, you get back to the same graph $$cos(t) = cos(-t) $$ $$ sin(t) = -sin(-t)$$
- $tan(t)$ is also an odd function $$ tan(t) = -tan(-t)$$ 
- Worth noting that their reciprocal counterparts are also of the same classification i.e. $cosec(t)$ is odd, $sec(t)$ is even and $cot(t)$ is odd
- Because these are periodic functions, they have a period of $2\pi$ meaning they repeat themselves every $2\pi$ distance (once again same with their reciprocal counterparts) $$ cos(t) = cos(2\pi+t) $$ $$ sin(t) = sin(2\pi+t) $$ $$ tan(t) = tan(2\pi+t) $$
- $tan(t)$ repeats more often with period $\pi$ $$tan(t)=tan(t+\pi)$$
- $cos(t)$ and $sin(t)$ flips with $\pi$ distance $$cos(t)=-cos(t+\pi) $$ $$ sin(t)=-sin(t+\pi)$$
- At origin $(0,0)$, both $sin(t)$ and $tan(t)$ crosses it but $cos(t)$ does not. Interestingly, non of the reciprocal functions touches 0 since you cannot divide by 0 $$sin(0) = 0 ;\space cos(0) = 1;\space tan(0) = 1$$
- At $\pi$ $$sin(\pi) = 0 ;\space cos(\pi) = -1;\space tan(\pi) = 0$$
- $tan(t)$ and the reciprocal counterparts all have asymptotes. The asymptotes exists where the original function crosses the x axis.

#### <a id='toc1_1_6_1_'></a>[Standard Trignometric Identities](#toc0_)

There are a few formulas to prove here... First to tackle is the sum of two angles identities and then the difference of two angles identities. We can use these proven identities to derive the double angle, half angle, sum-to-product and product-to-sum identities.

The majority of the work to prove the identies is done by extending the original triangle by draw a rectangle around it and then deriving the vertices in term of $cos$ and $sin$. It is useful to draw this out on paper when going through it.

##### <a id='toc1_1_6_1_1_'></a>[Sum of two angles formula](#toc0_)
 
$$ sin(y+x) = sin(y)cos(x) + sin(x)cos(y) $$
$$ cos(y+x) = cos(y)cos(x) - sin(x)sin(y) $$

From the diagram above, we can see that: $$ CB = DF = DE + EF $$ $$ DC = FB = FA + AB $$

We can derive that (because the hypotenuse is 1 on the unit circle) $$ BC = sin(x+y) ,\space AB = cos(x+y) $$

On the otherside of the hypotenuse $AC$ for triangle $\triangle EDA $, we have $$ EA = sin(x) ,\space EC = cos(x) $$ 

We can also observe $EC$ is the hypotenuse for the right angle triangle $\triangle EDC $ since $\square FBCD$ is a rectangle with 4 square angles. Therefore... 
$$ cos(y) = \frac{DC}{EC},\space sin(y) = \frac{DE}{EC}$$

Since $ EC = cos(x) $, we can replace the $EC$ above and rearrange to get: $$ cos(y) = \frac{DC}{cos(x)} \Rightarrow DC = cos(x)cos(y) $$  $$ sin(y) = \frac{DE}{cos(x)} \Rightarrow DE = cos(x)sin(y) $$

Lastly, we can see that $\angle FEA$ is the same as angle $y$ or $\angle ECD$ since $$ \angle ECD + \angle DEC + \angle EDC = 180^o $$ $$ \angle FEA = 180^o - \angle CEA - \angle DEC \Rightarrow \angle FEA = y $$ 

Since we know the hypotenuse $EA = sin(x)$ and angle $\angle FEA = y$, we can deduce the vertices $EF$ and $FA$ as $$ cos(y) = \frac{EF}{EA} =  \frac{EF}{sin(x)} \Rightarrow EF = cos(y)sin(x) $$ $$ sin(y) = \frac{FA}{EA} =  \frac{FA}{sin(x)} \Rightarrow FA = sin(y)sin(x) $$

Ok,  let's look at formulas we derived...

$$ AB = cos(x+y), \space FA = sin(y)sin(x) $$ $$ BC = sin(x+y) $$ $$ DC = cos(x)cos(y) $$ $$ DE = cos(x)sin(y), \space EF = cos(y)sin(x) $$

And finally putting it together, we can see the 2 formulas above:

$$ BC = DE + EF $$ $$ sin(x+y) = cos(x)sin(y) + cos(y)sin(x) $$ and  $$ DC = FA + AB \Rightarrow cos(x)cos(y) = cos(x+y) + sin(y)sin(x) $$ $$ cos(x+y) = cos(x)cos(y) - sin(y)sin(x) $$

Similar to above, $$AB = FD = FE + ED $$ $$ FA = DB = DC + BC $$

We can also derive for $\triangle ABC$ that $$ BC = sin(x-y),\space AB = cos(x-y) $$

On the otherside of the $AC$ hypotenuse for triangle $\triangle AEC$, we have $$ AE = cos(y),\space EC = sin(y) $$

Since now also have the hypotenuse $AE$ as well as angle $\angle AEF$ (same angle as $\angle EAB$ since they are supplementary angles made by parallel lines which adds up to $180^o \therefore$ $\angle AEF = x$), we can derive the other edges for the triangle $\triangle FAE$ $$ sin(x) = \frac{FA}{cos(y)} \Rightarrow FA = sin(x)cos(y) $$ $$ cos(x) = \frac{FE}{cos(y)} \Rightarrow FE = cos(x)cos(y) $$

The last triangle to look at is $\triangle EDC$. We know the hypotenuse is $EC$ which is $sin(y)$. Since we know that $\angle AEF = x \Rightarrow \angle ECD = x$ (should be fairly easy to prove why). With this we have $$ sin(x) = \frac{ED}{sin(y)} \Rightarrow ED = sin(x)sin(y) $$ $$ cos(x) = \frac{DC}{sin(y)} \Rightarrow DC = cos(x)sin(y) $$

Once again, let's assemble all the formulas we derived $$ AB = cos(x-y) $$ $$ FE = cos(x)cos(y),\space ED = sin(x)sin(y) $$ $$ FA = sin(x)cos(y) $$ $$ DC = cos(x)sin(y), \space BC = sin(x-y) $$

And then putting all together $$ AB = FE + ED $$ $$ cos(x-y) = cos(x)cos(y) + sin(x)sin(y) $$ 

and $$ FA = DC + BC \Rightarrow sin(x)cos(y) = cos(x)sin(y) + sin(x-y) $$ $$ sin(x-y) = sin(x)cos(y) - cos(x)sin(y) $$  


##### <a id='toc1_1_6_1_3_'></a>[Generalisation of sum and difference formulas & derivation of tan function](#toc0_)

With the formulas above, we can see $$cos(x \pm y) = cos(x)cos(y) \mp sin(x)sin(y) $$ $$ sin(x \pm y) = sin(x)cos(y) \pm sin(y)cos(x) $$

Using the results above, we can also derive the $tan(x+y)$ and $tan(x-y)$. $$ tan(x \pm y) = \frac{sin(x \pm y)}{cos(x \pm y)} = \frac{sin(x)cos(y) \pm sin(y)cos(x)}{cos(x)cos(y) \mp sin(x)sin(y)} $$ 

If we divide both numerator and denominator with $\frac{1}{cos(x)cos(y)}$, the equation simplifies down significantly $$ tan(x \pm y) = \frac{\frac{sin(x)cos(y)}{cos(x)cos(y)} \pm \frac{sin(y)cos(x)}{cos(x)cos(y)}}{\frac{cos(x)cos(y)}{cos(x)cos(y)} \mp \frac{sin(x)sin(y)}{cos(x)cos(y)}} $$ $$ tan(x \pm y) = \frac{tan(x) \pm tan(y)}{1 \mp tan(x)tan(y)} $$



##### <a id='toc1_1_6_1_4_'></a>[Double angle formula](#toc0_)

Using the sum of two angles formula, we can derive the double angle formula. Instead of $sin(x+y)$ or $cos(x+y)$, we can replace the y.

For sin function $$ sin(x+y) \rightarrow sin(x+x) = sin(2x) $$ $$ \Rightarrow sin(x)cos(x) + sin(x)cos(x) $$ $$ \Rightarrow 2sin(x)cos(x) $$

For cos function $$ sin(x+y) \rightarrow cos(x+x) = cos(2x) $$ $$ \Rightarrow cos(x)cos(x) - sin(x)sin(x) $$ $$ \Rightarrow cos^2(x) - sin^2(x) $$
 
For tan function, we can use the 2 formula we just derived $$ tan(2x) = \frac{sin(2x)}{cos(2x)} = \frac{2sin(x)cos(x)}{cos^2(x) - sin^2(x)} $$ 

We can divide numerator and denominator with $cos^2(x)$ $$ tan(2x) = \frac{ \frac{2sin(x)cos(x)}{cos^2(x)} }{\frac{(cos^2(x) - sin^2(x))}{{cos^2(x)}}} $$ $$ \Rightarrow \frac{2tan(x)}{1-tan^2(x)} $$

You can also get the same formula for $tan$ using the $tan(x+y)$ formula we derived previously

##### <a id='toc1_1_6_1_5_'></a>[Half angle formula](#toc0_)

From double angle, we can also have half angle formula. We know from above $$ cos^2(x) - sin^2(x) = cos(2x) $$ $$ cos^2(x) + sin^2(x) = 1 $$  

With some simple rearrangement of the second formula $$ cos^2(x) = 1 - sin^2(x) $$ $$ sin^2(x) = 1 - cos^2(x) $$  

Putting this back into the double angle formula $$ (1 - sin^2(x)) - sin^2(x) = cos(2x) \Rightarrow 1-2sin^2(x) = cos(2x)  $$ $$ cos^2(x) - (1 - cos^2(x)) = cos(2x) \Rightarrow 2cos^2(x) -1 =  cos(2x) $$

If we set $$\theta = 2x \Rightarrow \frac{\theta}{2} = x$$

And the substitute back in to the two formula we derived just above 

$$ 1-2sin^2(x) = cos(2x) \rightarrow 1-2sin^2(\frac{\theta}{2}) = cos(\theta) $$ $$ 2cos^2(x) - 1 =  cos(2x) \rightarrow 2cos^2(\frac{\theta}{2}) - 1 =  cos(\theta) $$ 


We now just need to rearrange the formulas so that the $\frac{\theta}{2}$ is one side.

For sin function

$$ 1-2sin^2(\frac{\theta}{2}) = cos(\theta) $$ $$ \Rightarrow 2sin^2(\frac{\theta}{2}) = 1 - cos(\theta) $$ $$ \Rightarrow sin^2(\frac{\theta}{2}) = \frac{1 - cos(\theta)}{2} $$ $$ \Rightarrow sin(\frac{\theta}{2}) = \sqrt{\frac{1 - cos(\theta)}{2}} $$

And for cos function

$$ 2cos^2(\frac{\theta}{2}) - 1 =  cos(\theta)  $$ $$ \Rightarrow 2cos^2(\frac{\theta}{2}) = 1 + cos(\theta) $$ $$ \Rightarrow cos^2(\frac{\theta}{2}) = \frac{1 + cos(\theta)}{2} $$ $$ \Rightarrow cos(\frac{\theta}{2}) = \sqrt{\frac{1 + cos(\theta)}{2}} $$

For tan function, we can once again derive using sin and cos function. Gonna use $x$ again because $\theta$ is a pain in the ass to write using latex $$ tan(\frac{x}{2}) = \frac{sin(\frac{x}{2})}{cos(\frac{x}{2})} = \frac{\sqrt{\frac{1 - cos(x)}{2}}}{\sqrt{\frac{1 + cos(\theta)}{2}}} $$ 

Cancel out the $\sqrt{\frac{1}{2}}$ on both the denominator and numerator, we get $$ tan(\frac{x}{2}) = \sqrt{\frac{1 - cos(x)}{1+cos(x)}} $$

Rationalise the denominator by multiplying both the denominator and numerator with $\sqrt{1-cos(x)}$ 

$$ \sqrt{\frac{(1 - cos(x))(1 - cos(x))}{(1+cos(x))(1 - cos(x))}} = \sqrt{\frac{(1 - cos(x))^2}{1^2 - cos^2(x)}} = \frac{1-cos(x)}{\sqrt{sin^2(x)}} $$ $$ \Rightarrow tan(\frac{x}{2}) = \frac{1-cos(x)}{sin(x)}  $$


##### <a id='toc1_1_6_1_6_'></a>[Product-to-Sum formula](#toc0_)

Recall from the above, we derived the sum and difference of 2 angles formula

$$cos(x \pm y) = cos(x)cos(y) \mp sin(x)sin(y) $$ $$ sin(x \pm y) = sin(x)cos(y) \pm sin(y)cos(x) $$

The product-to-sum formulas flips the formula and makes the products (e.g. $cos(x)cos(y) = ...$, $sin(x)sin(y) = ...$, $sin(x)cos(y) = ...$, $sin(y)cos(x) = ...$) the left handside while the sum is on the right handside.

Let's take a look at the $cos(x)cos(y)$ and $sin(x)sin(y)$ formulas which relies on the $cos(x \pm y)$ formula, we will need to isolate the products from the original equation. We can do this by adding/subtracting $cos(x \pm y)$ against itself.

For $cos(x)cos(y)$
$$ cos(x+y) + cos(x-y) = (cos(x)cos(y) - sin(x)sin(y)) + (cos(x)cos(y) + sin(x)sin(y)) $$
$$ = cos(x)cos(y) - sin(x)sin(y) + cos(x)cos(y) + sin(x)sin(y) $$
$$ \Rightarrow cos(x+y) + cos(x-y) = 2cos(x)cos(y) $$
$$ \Rightarrow cos(x)cos(y) = \frac{cos(x+y) + cos(x-y)}{2}  $$

For $sin(x)sin(y)$
$$ cos(x+y) - cos(x-y) = (cos(x)cos(y) - sin(x)sin(y)) - (cos(x)cos(y) + sin(x)sin(y)) $$
$$ = cos(x)cos(y) - sin(x)sin(y) - cos(x)cos(y) - sin(x)sin(y) $$
$$ \Rightarrow cos(x+y) - cos(x-y) = -2sin(x)sin(y) $$
$$ \Rightarrow sin(x)sin(y) = \frac{cos(x-y) - cos(x+y)}{2}  $$

We can move to $sin(x)cos(y)$ and $sin(y)cos(x)$ which we will use $sin(x \pm y)$ instead but follow the same strategy

For $sin(x)cos(y)$
$$ sin(x+y) + sin(x-y) = (sin(x)cos(y) + sin(y)cos(x)) + (sin(x)cos(y) - sin(y)cos(x)) $$
$$ = sin(x)cos(y) + sin(y)cos(x) + sin(x)cos(y) - sin(y)cos(x) $$
$$ \Rightarrow sin(x+y) + sin(x-y) = 2sin(x)cos(y) $$
$$ \Rightarrow sin(x)cos(y) = \frac{sin(x+y) + sin(x-y)}{2}  $$

For $sin(y)cos(x)$
$$ sin(x+y) - sin(x-y) = (sin(x)cos(y) + sin(y)cos(x)) - (sin(x)cos(y) - sin(y)cos(x)) $$
$$ = sin(x)cos(y) + sin(y)cos(x) - sin(x)cos(y) + sin(y)cos(x) $$
$$ \Rightarrow sin(x+y) - sin(x-y) = 2sin(y)cos(x) $$
$$ \Rightarrow sin(y)cos(x) = \frac{sin(x+y) - sin(x-y)}{2}  $$

##### <a id='toc1_1_6_1_7_'></a>[Sum-to-Product formula](#toc0_)

We can use the product-to-sum formulas to get the sum-to-product formulas.

This now means making the sum (e.g. $sin(x) + sin(y) =...$, $cos(x) + cos(y) =...$, $sin(x) - sin(y) =...$, $cos(x) - cos(y) =...$). 

Notice that these coincides with the way how the product-to-sum formulas laid out except that instead of $sin(x) + sin(y)$, it is $sin(x+y) + sin(x-y)$. Let's use some dummy variables ($\alpha$ and $\beta$) to replace the sum and differences.

Let $$x+y = \alpha$$ $$x-y = \beta$$

Thus in term of $\alpha$ and $\beta$, we have $x$ as $$ (x+y) + (x-y) = \alpha + \beta $$ $$ 2x = \alpha + \beta $$ $$ \Rightarrow x = \frac{\alpha + \beta}{2} $$ 

and $y$ as $$ (x+y) - (x-y) = \alpha - \beta $$ $$ 2y = \alpha - \beta $$ $$ \Rightarrow y = \frac{\alpha-\beta}{2} $$ 

Replacing this back into the product-to-sum formulas, we get the below results (just replace the dummy $\alpha$ and $\beta$ variables with $x$ and $y$ respectively at the end) 

For $cos(x) + cos(y)$ $$ cos(x)cos(y) = \frac{cos(x+y) + cos(x-y)}{2} $$ $$ \Rightarrow cos(\frac{\alpha + \beta}{2})cos(\frac{\alpha-\beta}{2}) = \frac{cos(\alpha) + cos(\beta)}{2} $$ $$ \Rightarrow cos(\alpha) + cos(\beta) = 2cos(\frac{\alpha + \beta}{2})cos(\frac{\alpha-\beta}{2}) $$ 

For $cos(x) - cos(y)$ $$ sin(x)sin(y) = \frac{cos(x-y) - cos(x+y)}{2} $$ $$ \Rightarrow sin(\frac{\alpha +\beta}{2})sin(\frac{\alpha-\beta}{2}) = \frac{cos(\beta) - cos(\alpha)}{2} $$ $$ \Rightarrow cos(\beta) - cos(\alpha) = 2sin(\frac{\alpha + \beta}{2})sin(\frac{\alpha-\beta}{2}) $$ $$ \Rightarrow cos(\alpha) - cos(\beta) = -2sin(\frac{\alpha + \beta}{2})sin(\frac{\alpha-\beta}{2}) $$ 

For $sin(x) + sin(y)$ 
$$ sin(x)cos(y) = \frac{sin(x+y) + sin(x-y)}{2} $$ $$ \Rightarrow sin(\frac{\alpha +\beta}{2})cos(\frac{\alpha-\beta}{2}) = \frac{sin(\alpha) + sin(\beta)}{2} $$ $$ \Rightarrow sin(\alpha) + sin(\beta) = 2sin(\frac{\alpha + \beta}{2})cos(\frac{\alpha-\beta}{2}) $$ 

For $sin(x) - sin(y)$
$$ sin(y)cos(x) = \frac{sin(x+y) - sin(x-y)}{2} $$ $$ \Rightarrow sin(\frac{\alpha -\beta}{2})cos(\frac{\alpha+\beta}{2}) = \frac{sin(\alpha) - sin(\beta)}{2} $$ $$ \Rightarrow sin(\alpha) - sin(\beta) = 2sin(\frac{\alpha - \beta}{2})cos(\frac{\alpha+\beta}{2}) $$

### <a id='toc1_1_7_'></a>[Composite & Inverse Functions](#toc0_)

#### <a id='toc1_1_7_1_'></a>[Composite functions](#toc0_)
Composite function is a function that is composed for 2 more functions being applied consecutively. 

An example is where you have 2 functions $$f(x) = x^2 $$ $$ g(x) = sin(x)$$ 

We can define the composition function $f \circ g$ by first applying $g$ and then $f$. This DOES NOT mean the two function are multiplied but rather it is a nested function $$f \circ g = f(g(x))$$

Substitute the definition of $f$ and $g$ back in, we get $$ f(g(x)) = f(sin(x)) = sin^2(x) $$

It's worth noting that you apply the function matters and, in general, the reverse will not give the same result $$f \circ g \ne g \circ f \Rightarrow f(g(x)) \ne g(f(x)) $$ $$ sin^2(x) \ne sin(x^2) $$ 

#### <a id='toc1_1_7_2_'></a>[Inverse function](#toc0_)

A function is can thought of as, for a normal function $f$, you put in $x$ and you get $f(x)$ or $y$ (or given $x$, what is $y$?) 

$$ x \rightarrow \boxed{f} \rightarrow f(x) \space or \space y $$

An inverse function is effectively undoes what $f(x)$ does, how can you go to $x$ from $y$ (or given $y$, what is $x$?) and can visualised like this

$$ x \leftarrow \boxed{f^{-1}}  \leftarrow f(x) \space or \space y $$

To create the inverse function, we would then need to make $y = f(x)$ and then solve for $x$. Here's an example $$ f(x) = y = 3x - 4 $$

We rearrange to solve for x by $$ y = 3x - 4 \Rightarrow x =  \frac{y + 4}{3} \Rightarrow f^{-1} = \frac{y + 4}{3} $$

Note that not all functions have an inverse function. An example is quadratic equation $y=x^2$. Whilst every value of $x$ or the domain of $x$ maps to a value of $y$, a single value of $y$ can map to two values of $x$. Thus the below condition must be met in order for a function to have an inverse function $$ f(f^{-1}(x)) = x $$ 

or $$ f^{-1}(f(x)) = x $$ 

You would have notice that condition for a function to have an inverse function is the special case where the composite function order does not matter.

However, if we bound the domain of $f(x) =x^2$ to $(0, \infty)$, this now restrict the mappings so that only one value of $x$ will map to $y$. Thus $f(x) =x^2$ will have a domain $(0, \infty)$ and range $(0, \infty)$ and $f^{-1}(x) = +\sqrt{y}$

Notice due to the periodic nature of the trignometric functions, their domain also need to be bounded. 

- For $sin(x)$, domain $[-\frac{\pi}{2}, \frac{\pi}{2}]$ would result in range $[1,1]$ for $sin^{-1}(x)$
- For $cos(x)$, domain $[0, \pi]$ would result in range $[1,1]$ for $cos^{-1}(x)$
- For $tan(x)$, domain $[-\frac{\pi}{2}, \frac{\pi}{2}]$ would result in range $(-\infty,\infty)$ for $tan^{-1}(x)$

Lastly, the graph of the inverse function is the reflection of the function at $y = x$. Took me a while to understand why this was the case but a little proof helped. 

Suppose point $D_0(x,y)$ is a point on the graph plotted by $f(x)$ since y = f(x)$. Same thing with the graph plotted using x = f^{-1}(y), we have point $D_1(y,x)$. 

Two points are symmetrical w.r.t to a line $a$ if the line $a$ is perpendicular to the line $b$ created by connecting $D_0(x,y)$ and $D_1(y,x)$ at its midpoint $M$. 

If we set the line $a$ to be $x=y$, we need to prove that basically the midpoint $M$ lies on $x=y$ and line $b$ is perpendicular to line $a$. 

The midpoint $M$ is $\frac{x+y}{2}, \frac{x+y}{2}$, since both the $x$ and $y$ coordinate are the same, midpoint $M$ lies on the $x=y$ line.

We can check if 2 lines are perpendicular if the product of their slope is $-1$. We know the gradientor $x=y$ is 1 by definition. The gradient for line $b$ is $$ \frac{x-y}{y-x} \Rightarrow -\frac{x-y}{x-y} = -1 $$ 

We can see that the product of the two gradient is -1 and the line intersect at midpoint $M$ and thus the graph of the inverse function is the reflection of the function at $y = x$

### <a id='toc1_1_8_'></a>[Indicator Functions](#toc0_)

For a given set $X$ with a subset $A$, the indicator function $I$ for $A$ is a function from $X$ to {1,0} (1 or 0) such that $$ I_A(x) = \begin{cases} 1 & x \in A \\ 0 & x \notin A \end{cases}$$

Given 2 sets, A and B, the indicator function $I_{A \cap B}(x)$ of the intersection $A \cap B $ will be 1 if x is in both $I_A(x) = 1$ and $I_B(x) = 1$. 

If $I_A(x) = 0$ or $I_B(x) = 0$, then the indicator function $I_{A \cap B}(x) = 0$. We can write $$ I_{A \cap B}(x) = I_A(x) I_B(x) $$

We can thus reason that the union will be $$ I_{A \cup B}(x) = I_A(x) + I_B(x) - I_A(x) I_B(x) $$

This is because $I_A(x) = 1$ when $x$ is in $A$ and likewise for $B$, $I_B(x) = 1$. The union between them is $I_A(x) + I_B(x)$ but this is overlapped at the intersection which we will need to deduct like above.
